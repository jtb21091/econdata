```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'africa.xlsx'  # Replace with the actual path
data = pd.read_excel(file_path)

# Calculate percentage change from 2022 to 2023
data['% Change 2022-2023'] = ((data['TOTAL 2023'] - data['TOTAL 2022']) / data['TOTAL 2022']) * 100

# Calculate percentage change from 2023 to 2024 YTD
data['% Change 2023-2024 YTD'] = ((data['TOTAL 2024 YTD'] - data['TOTAL 2023 YTD']) / data['TOTAL 2023 YTD']) * 100

# Define the output file path
output_file_path = 'percentage_changes.xlsx'  # Replace with the desired output path

# Export the results to a new Excel file
data.to_excel(output_file_path, index=False, sheet_name='Percentage Changes')

# Sort the data by percentage change
data_sorted_2022_2023 = data.sort_values(by='% Change 2022-2023')
data_sorted_2023_2024 = data.sort_values(by='% Change 2023-2024 YTD')

# Plot the percentage changes
plt.figure(figsize=(14, 7))

# Plot percentage change from 2022 to 2023
plt.subplot(1, 2, 1)
plt.bar(data_sorted_2022_2023['Country'], data_sorted_2022_2023['% Change 2022-2023'], color='blue')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('% Change')
plt.title('% Change 2022-2023 by Country')

# Plot percentage change from 2023 to 2024 YTD
plt.subplot(1, 2, 2)
plt.bar(data_sorted_2023_2024['Country'], data_sorted_2023_2024['% Change 2023-2024 YTD'], color='green')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('% Change')
plt.title('% Change 2023-2024 YTD by Country')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()

print(f"Data has been successfully exported to {output_file_path}")
```
