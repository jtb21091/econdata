```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'SCM GHG 2017 NAICS code.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Limit the data to top 25 emissions
top_25_data = df.nlargest(25, 'Supply Chain Emission Factors with Margins')

# Plotting the bar chart for top 25 emissions
plt.figure(figsize=(12, 8))
plt.bar(top_25_data['2017 NAICS Title'], top_25_data['Supply Chain Emission Factors with Margins'], color='skyblue')
plt.xlabel('NAICS Title')
plt.ylabel('Emission Factors')
plt.title('Top 25 Supply Chain Emission Factors with Margins for 2017 NAICS Codes')
plt.xticks(rotation=90, ha='center')
plt.tight_layout()
plt.show()
```
