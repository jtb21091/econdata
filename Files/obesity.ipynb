```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'LakeCounty_Health_2397514566901885190.csv'
data = pd.read_csv(file_path)

# Sort the data by Obesity
sorted_data = data.sort_values(by='Obesity')

# Calculate the average obesity rate
average_obesity = sorted_data['Obesity'].mean()

# Create the plot
plt.figure(figsize=(14, 8))
bars = plt.bar(sorted_data['NAME'], sorted_data['Obesity'], color='skyblue')

# Highlight bars above average
for bar in bars:
    if bar.get_height() > average_obesity:
        bar.set_color('salmon')

# Add average line
plt.axhline(average_obesity, color='gray', linestyle='--')
plt.text(len(sorted_data)-1, average_obesity + 0.5, f'Average: {average_obesity:.2f}%', color='gray', ha='center')

# Add labels and title
plt.xlabel('State')
plt.ylabel('Obesity Rate (%)')
plt.title('Obesity Rates by State')
plt.xticks(rotation=90, ha='center')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```
