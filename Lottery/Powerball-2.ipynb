```python
import pandas as pd

# Load the data
file_path = 'Lottery_Powerball_Winning_Numbers__Beginning_2010.csv'
df = pd.read_csv(file_path)

# Extract the winning numbers
winning_numbers = df['Winning Numbers'].str.split(' ', expand=True)
winning_numbers = winning_numbers.apply(pd.to_numeric, errors='coerce')

# Stack the numbers into a single series and count occurrences
number_counts = winning_numbers.stack().value_counts()

# Get the 6 most popular numbers
top_6_numbers = number_counts.head(6)

print("Top 6 Most Popular Numbers:")
print(top_6_numbers)
```

```python
import matplotlib.pyplot as plt

# Get the top 69 numbers
top_69_numbers = number_counts.head(69)

# Plot the top 50 numbers
plt.figure(figsize=(12, 8))
top_69_numbers.plot(kind='bar', color='skyblue')
plt.title('Top 69 Most Popular Powerball Numbers')
plt.xlabel('Powerball Numbers')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
