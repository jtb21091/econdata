```python
import pandas as pd

# Load the Excel file
file_path = 'GDP per Capita NAFTA countries.xlsx'
data = pd.read_excel(file_path, sheet_name=None)

# Display the sheet names to understand the structure of the file
data.keys()
```

```python
# Load the data from the sheet
df = data['Sheet1']

# Display the first few rows of the dataframe to understand its structure
df.head()
```

```python
# Cleaning up the dataframe
df.columns = df.iloc[0]
df = df.drop(0)
df = df.reset_index(drop=True)

# Convert the columns to appropriate data types
df['Year'] = df['Year'].astype(int)
df['USA'] = df['USA'].astype(float)
df['Canada'] = df['Canada'].astype(float)
df['Mexico'] = df['Mexico'].astype(float)

# Display the cleaned dataframe
df.head()
```

```python
import matplotlib.pyplot as plt

# Plotting the GDP per Capita for all three countries
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['USA'], label='USA')
plt.plot(df['Year'], df['Canada'], label='Canada')
plt.plot(df['Year'], df['Mexico'], label='Mexico')
plt.xlabel('Year')
plt.ylabel('GDP per Capita')
plt.title('GDP per Capita for NAFTA Countries (1994-2023)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the total percentage change for each country
usa_change = ((df['USA'].iloc[-1] - df['USA'].iloc[0]) / df['USA'].iloc[0]) * 100
canada_change = ((df['Canada'].iloc[-1] - df['Canada'].iloc[0]) / df['Canada'].iloc[0]) * 100
mexico_change = ((df['Mexico'].iloc[-1] - df['Mexico'].iloc[0]) / df['Mexico'].iloc[0]) * 100

# Creating a bar chart for the total percentage change
countries = ['USA', 'Canada', 'Mexico']
percentage_changes = [usa_change, canada_change, mexico_change]

plt.figure(figsize=(8, 5))
plt.bar(countries, percentage_changes, color=['blue', 'red', 'green'])
plt.xlabel('Country')
plt.ylabel('Total Percentage Change (%)')
plt.title('Total Percentage Change in GDP per Capita (1994-2023)')
plt.show()
```
