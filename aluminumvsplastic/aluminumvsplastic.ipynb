```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the graph
materials = ['Aluminum', 'HDPE Plastic']
prices = [2564.54, 1400]  # in USD per metric ton
ghg_emissions = [16.5, 2.2]  # in metric tons of CO2e per metric ton

# Creating subplots with data labels
fig, ax1 = plt.subplots()

# Plotting price per metric ton
ax1.set_xlabel('Material')
ax1.set_ylabel('Price (USD per metric ton)', color='tab:blue')
bars = ax1.bar(materials, prices, color='tab:blue', alpha=0.6, label='Price')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Adding data labels for price
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='tab:blue')

# Creating a second y-axis to plot GHG emissions
ax2 = ax1.twinx()
ax2.set_ylabel('GHG Emissions (metric tons of CO2e per metric ton)', color='tab:green')
line, = ax2.plot(materials, ghg_emissions, color='tab:green', marker='o', label='GHG Emissions')
ax2.tick_params(axis='y', labelcolor='tab:green')

# Adding data labels for GHG emissions
for i, txt in enumerate(ghg_emissions):
    ax2.annotate(txt, (materials[i], ghg_emissions[i]), textcoords="offset points", xytext=(0,10), ha='center', color='tab:green')

# Adding titles and labels
plt.title('Aluminum vs Plastic: Price and GHG Emissions per Metric Ton')
fig.tight_layout()

# Adding legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Displaying the plot
plt.show()
```
